```python
#set subsample = sorted_rel_dif in the above code (commented out) to see how decoding changes when removing reliable neurons
 
def reliability(all_run_tensors, all_stat_tensors):
    diff_reliability = []
    rel_run = []    
    rel_stat = []
    
    mean_var_run = []
    mean_var_stat = []
    
    all_var_run = []
    all_var_stat = []

    sort_inds_dif = {}
    keys = []
    
    for key in sorted(all_run_tensors.keys()):


        data_run, trialNum_run = all_run_tensors[key]
        data_stat, trialNum_stat = all_stat_tensors[key]
        
  
        run_data, run_labels, stat_data, stat_labels = tensorize2(np.array(data_run),
                                                                  np.array(trialNum_run), 
                                                                  np.array(data_stat),
                                                                  np.array(trialNum_stat), 
                                                                  balance_classes = True)
    
    
    
    
        n_classes = len(np.unique(run_labels))

        ten_run = []
        ten_stat = []
        for i in range(n_classes):

            ten_run.append(np.array(run_data)[np.array(run_labels) == i])
            ten_stat.append(np.array(stat_data)[np.array(stat_labels) == i])



        reliability_run = calculate_reliability(ten_run)
        reliability_stat = calculate_reliability(ten_stat)
        
        var_av_r, baseline_r = calculate_meanvar(ten_run)
        var_av_s, baseline_s = calculate_meanvar(ten_stat)

        all_var_run.append(baseline_r)#(baseline_r))
        all_var_stat.append(baseline_s)#(baseline_s))

        mean_var_run.append((var_av_r))
        mean_var_stat.append((var_av_s))


        diff_reliability.append([(2*(np.average(reliability_run) - np.average(reliability_stat)))/ 
                                 (np.average(reliability_run) + np.average(reliability_stat))]*len(var_av_r))



        per_neuron_rel = np.array(2*(reliability_run - reliability_stat)/ (reliability_run + reliability_stat))
        #diff_reliability.append(per_neuron_rel)
        
        
        #sorts in ascending order

        n_examples, n_neurons  = np.array(run_data).shape
        
        exclude_num = int(np.ceil(n_neurons * .25))
        sorted_rel_dif = np.argsort(per_neuron_rel)[:-exclude_num]
        
        sort_inds_dif[key] = sorted_rel_dif#sorted_rel_dif[:-exclude_num]

        rel_run.append(reliability_run)
        rel_stat.append(reliability_stat)

        keys.append([key]*len(var_av_r))
        
    return diff_reliability, rel_run, rel_stat, all_var_run, all_var_stat, mean_var_run, mean_var_stat, sort_inds_dif, keys
```
